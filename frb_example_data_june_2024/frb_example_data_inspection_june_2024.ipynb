Lets make a dict to map the FRB names to their data

```python
# make a dict? mapping the Burstname collum to the loaded in .npy files
# actually we are going to be doing this with the BurstNames col

burst_names = burst_parameters.BurstName.to_list()

# Dictionary to store BurstName -> data mapping
burst_data_dict = {}

# load in files
# make dict mapping burst names (key) to data (value)

for burst_name in burst_names:
    file_name = f"{burst_name}_burst_data.npy"
    try:
        # Load the .npy file
        data = np.load(file_name)
        # Map BurstName to loaded data
        burst_data_dict[burst_name] = data
    except FileNotFoundError:
        print(f"Warning: File {file_name} not found.")



```

```python
# appears to be 2D, o maybe this is a 'waterfall plot' data?
# great, it is a waterfall plot!

def waterfall_plot(frb_name, save=False, show=True):
    '''makes a waterfall plot for frbbdata from the frb with frb_name
    I am assuming its CHIME data and making the freq axis correspond to the full 400-800 Mhz range
    '''
    frb_data = burst_data_dict[frb_name]


    # physical y units 
    index_to_Mhz = 400 / frb_data.shape[0] 
    # CHIME bandwidth is 400-800 Mghz , so each channel is 400 / total channels 
    # Determine y tick locations and labels
    num_ticks = 5
    y_ticks_locations = np.linspace(0, frb_data.shape[0] - 1, num_ticks,) 
    y_ticks_labels = np.round(np.linspace(0, frb_data.shape[0] - 1, num_ticks, dtype=int) * -1*index_to_Mhz + 800, 
                               decimals=-2).astype(int)
    # convert index to freq ^^ 

     
  

    # lets make this look pretty
    plt.figure(figsize=(8,8), dpi=100)
    plt.imshow(frb_data,)
    # Calculate physical units for y ticks
    y_ticks_physical = np.arange(5) * index_to_Mhz 
    
    plt.xlabel('Time Units =?', fontsize=15)

    plt.yticks(y_ticks_locations, y_ticks_labels, ) #physical units 
    plt.ylabel('Freq (MHz)', fontsize=15)  # Label for y-axis

    plt.title(frb_name + '\n CHIME Data', fontsize=15)

    colorbar = plt.colorbar()
    colorbar.set_label('Intensity Units=? Maybe SNR?', fontsize=14)

    if save:
        plt.savefig('../../' + frb_name + 'waterfall', bbox_inches='tight')

    if show:
        plt.show()
    if show != True:
        plt.close()
```

```python
waterfall_plot(frb_name, save=True)
```
